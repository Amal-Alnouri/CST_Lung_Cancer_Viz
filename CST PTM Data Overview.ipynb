# CST PTM Data Overview

The PTM data from CST has a significant amount of missing data and requires special consideration when normalizing. My starting point will be ratio-level-data where log2 ratios have been calculated from the cancerous cell lines compared to the non-cancerous 'Normal Pool' data from within the plex. 

This data is under the lung_cellline_3_1_16 directory and each PTM type has its own 'combined_ratios.tsv' file

I'll start by having a look at the phosphorylation data.
