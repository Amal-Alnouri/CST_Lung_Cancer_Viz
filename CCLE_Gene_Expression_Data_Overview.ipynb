# CCLE Gene Expression Data Overview 

The lung cancer cell line gene expression data was obtained from the Cancer Cell Line Encyclopedia [(CCLE)](https://portals.broadinstitute.org/ccle/home). Gene expression values were z-scored across all 1047 cancer cell lines in the CCLE. 37 of the 42 cell lines that were assayed by CST were also assayed by the CCLE. 

# Filtering for Genes of Interest
We defined genes of interest based on two criteria
* genes that were highly differentially expressed in particular cell lines
* genes that were highly expressed in many cell lines

In order to capture genes with both of these properties we filtered for genes that: 1) were differentially expressed with a Z-score of absolute value at or above 3 in at least 4 cell lines, 2) we allowed users to filter for the top genes based on the sum of their gene expression across all cell lines (using filter sliders). This filtering leaves us with 215 genes of interest. 

The interactive visualization can be seen here: [CST_Lung_Cancer_Viz](http://maayanlab.net/CST_Lung_Cancer_Viz/).

Several of the most-differentially expressed genes are known to be important in lung cancer and we will go over a few here.
